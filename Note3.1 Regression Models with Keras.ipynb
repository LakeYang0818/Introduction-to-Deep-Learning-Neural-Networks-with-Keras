### Step 3. Train/Test Split: split data into predictors and target


```python
concrete_data_columns = concrete_data.columns

# Predictors: all columns except for "Strength"
predictors = concrete_data[concrete_data_columns[concrete_data_columns != "Strength"]]

# Target: the "Strength" column
target = concrete_data["Strength"]
```

```python
predictors.head()
```

```python
target.head()
```

### Step 4. Data Preprocessing - Normalization

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
# save the number of predictors to n_cols to help build the network
n_cols = predictors_norm.shape[1]
```

### Step 5. Import Keras and Build a Neural Network

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# define the regression model
def regression_model():
    # create model
    ## 2 hidden layers, each of which has 50 hidden units (neurons)
    model = Sequential()
    model.add(Dense(50, activation = "relu", input_shape = (n_cols, )))
    model.add(Dense(50, activation = "relu"))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer = "adam", loss = "mean_squared_error")
    return model
```

### Step 6. Train and Test the Network

```python
# build the model
model = regression_model()

# fit the model
### 30% of data: validation
### Train for 100 epochs
model.fit(predictors_norm, 
          target, 
          validation_split = 0.3, 
          epochs = 100, 
          verbose = 2)
```
